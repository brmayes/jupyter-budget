```python
# Functions
def negativeNum(num):
    if (num > 0):
        return -num
    else:
        return num
    
def getMonth(date):
    if (date[:1] == '6'):
        return 'June'
    elif (date[:1] == '7'):
        return 'July'
    else:
        return 'August'
```

```python
# Import data
wells_fargo = pd.read_csv("transactions/wells_fargo.csv")
wells_fargo_cc = pd.read_csv("transactions/wells_fargo_cc.csv")
navy_fed = pd.read_csv("transactions/navy_fed.csv")
chase = pd.read_csv("transactions/chase.csv")
```

```python
# Combine dataframes into one 
frames = [wells_fargo,wells_fargo_cc,navy_fed,chase]
combined = pd.concat(frames)
df = combined.reset_index().drop('index',axis=1)
```

```python
# Apply negatives to all charges
df['Charge'] = df['Charge'].apply(negativeNum)
df.to_csv('all_transations.csv')
df
```

```python
# group by category
categories = df.groupby('Category').sum().reset_index()
categories.sort_values('Charge')
```
